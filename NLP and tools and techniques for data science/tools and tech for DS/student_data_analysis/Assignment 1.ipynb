## Adding new Column Grade Letter Task 2

```python
df['grade_letter'] = ''
df.loc[df['Marks']>=90,'grade_letter'] = 'A'
df.loc[(df['Marks']>=80) & (df['Marks']<90), 'grade_letter'] = 'B'
df.loc[(df['Marks']>=70) & (df['Marks']<80), 'grade_letter'] = 'C'
df.loc[(df['Marks']>=60) & (df['Marks']<70), 'grade_letter'] = 'D'
df.loc[df['Marks']<60, 'grade_letter'] = 'F'

```

## Students Grade +2 to new grade Task 5

```python
df.loc[df['Index']=='Std04', 'Marks'] = df.loc[df['Index']=='Std04', 'Marks'].replace(78, 78+2)
df.loc[df['Index']=='Std04', 'grade_letter'] = df.loc[df['Index']=='Std04', 'grade_letter'].replace('C', 'B')
```
